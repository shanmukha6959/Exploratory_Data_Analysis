### Load Dataset

```python
import pandas as pd
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
wine.data.shape
```

```python
#Independent varialbes or observations
wine.data
```

```python
#column names
wine.feature_names
```

```python
#Dependent variable
wine.target
```

```python
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df.head()
```

```python
#Add targets as last column in the DataFrame
df['target'] = wine.target
```

```python
df.head()
```
