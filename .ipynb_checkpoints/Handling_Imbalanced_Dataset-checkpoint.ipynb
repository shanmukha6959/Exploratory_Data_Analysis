### Handling_Imbalanced_Dataset

##### What is Imbalanced Dataset?

- When ever we are working on a data science project specifically on the classification problem. We will have the number of records or images and will have different targets or classes inside of our dataset. When we have `uneven distribution` of number of records for each class or target in our dataset.

- Eg. Anomoly detection problems most of the datasets are imbalanced, Fraud transaction detection, natural disaster.

- In Fraud detection, no fraud transactions are 95 and fraud transactions are 5 out of total 100 transactions. This particular data is imbalanced data.

##### What is the Impact of Imbalanced Dataset?

- When we are training a model or classifier using imbalanced dataset our training model may get more biased towards the majority class.

- Our model may not perform well on the class which is having less number of observations.

##### Techniques to handle Imbalanced dataset

For detecting fraud we may not only use accuracy but also use `right metrics` like `Accuracy, Precision, recall, F1-score` would be helpful in the `classification problem` when we have `imbalanced dataset`.



Another Technique:



    
- **1. UnderSampling**
    
    We take number of records in the minority class and match with the number of records(randomly choosen) in the majority class. 
    
    - Advantages
        - Reduce dataset size.
        - Low storage requirement.
        - Saves computation Cost.
        - Requires less run time.
        
    - Disadvantages
        - Will remove useful information while undersampling.
        - Random sampling may be biased sample.
        
    - When to use
        - Whe we have `good number of records in our minority class`.
   
   
- **2. OverSampling**
    
    We take the records in the minority class and increase the records by duplicating the records to match with the total records in the majority class. 
    
    - Advantages
        - No information loss like undersampling
        - Performs better than undersampling.
        
    - Disadvantages
        - likelyhood of overfitting as we duplicate the minority class observations.     
        
    - When to use
        - Whe we have too `less observation in minority class`.

- **3. SMOTE (Synthetic Minority Oversampling Technique)**

    SMOTE first finds all the closest nearest neighbors and creates new neighbors around the exisiting neighbors using KNN. It works on the KNN algorithms.
     
    - Advantages
        - Minimize the problem of overfitting caused by oversampling
        - No loss of useful information
        
    - Disdvantages
        - May introduce additional noise in data
        - Not effective for high dimensional data
    
    - When to use
        - When we have too less observation in minority class
