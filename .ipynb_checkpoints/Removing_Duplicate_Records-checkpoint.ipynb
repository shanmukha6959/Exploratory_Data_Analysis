### Removing duplicate records from dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("Dataset/automobile_data.csv")
df
```

#### drop duplicates

```python
df.drop_duplicates()
```

#### drop duplicates of every company

```python
df.drop_duplicates(subset = ['company'])
```

#### drop duplicates by keeping unique values of fuel_type and body_style

```python
df.drop_duplicates(subset = ['fuel_type', 'body_style'])
```

#### using keep argument we can keep last records while droppping the duplicates.

```python
df.drop_duplicates(subset = ['fuel_type', 'body_style'], keep = 'last')
```
